```python
# Get graph
g = Graph.from_json('ChainUtilityData/2018_all_votes_current.json')

# get populations
pops = np.array([g.nodes[n]['POP100'] for n in g.nodes])

# Set all nans to zero
pops = np.nan_to_num(pops)

# Save pops
np.save('ChainUtilityData/2018_all_votes_current_pop.npy', pops)

# update graph
for i,n in enumerate(g.nodes):
    
    
    if g.nodes[n]['POP100'] != g.nodes[n]['POP100']:
        g.nodes[n]['POP100'] = 0 
    
    if g.nodes[n]['SEN_DEM'] != g.nodes[n]['SEN_DEM']:
        g.nodes[n]['SEN_DEM'] = 0 
        
    if g.nodes[n]['SEN_REP'] != g.nodes[n]['SEN_REP']:
        g.nodes[n]['SEN_REP'] = 0 
        
    if g.nodes[n]['G_DEM'] != g.nodes[n]['G_DEM']:
        g.nodes[n]['G_DEM'] = 0 
        
    if g.nodes[n]['G_REP'] != g.nodes[n]['G_REP']:
        g.nodes[n]['G_REP'] = 0 
        
    g.nodes[n]['COMB_REP'] = 0.5*(g.nodes[n]['G_REP']+ g.nodes[n]['SEN_REP'])
    g.nodes[n]['COMB_DEM'] = 0.5*(g.nodes[n]['G_DEM']+ g.nodes[n]['SEN_DEM'])
    g.nodes[n]['US_Distric'] = g.nodes[n]['US_Distric'] - 1
    
# resave graph
g.to_json('ChainUtilityData/2018_all_votes_current_no_nan.json')
```
