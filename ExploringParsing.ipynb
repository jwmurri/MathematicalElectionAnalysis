```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('2020UtahPrimary/format2/2020 Primary Election - Grand County Precinct Level Results.csv')
```

```python
data.drop(['COUNTY NUMBER', 'PRECINCT CODE','Unnamed: 16'], axis=1, inplace=True)
#data.iloc[:,:6].drop([0,1], axis=0)
new_columns = list(data.columns)
new_columns[8:] = data.iloc[1, 8:]
data.columns = new_columns
data.drop([0,1], axis=0, inplace=True)
data.index = pd.Index(data['PRECINCT NAME'])
data.drop('PRECINCT NAME', axis=1, inplace=True)
data = data.astype(np.int32)

assert all(data[:-1].sum() == data.loc['COUNTY TOTALS', :])
data.drop('COUNTY TOTALS', axis=0, inplace=True)

assert all(data['BALLOTS CAST TOTAL'] >= data[['THOMAS E. WRIGHT, ROB BISHOP', 'SPENCER J. COX, DEIDRE M. HENDERSON', 'JON HUNTSMAN JR., MICHELLE KAUFUSI', 'GREG HUGHES, VICTOR IVERSON']].sum(axis=1))
assert all(data['BALLOTS CAST TOTAL'] >= data[['SEAN D. REYES', 'DAVID O. LEAVITT']].sum(axis=1))
```
