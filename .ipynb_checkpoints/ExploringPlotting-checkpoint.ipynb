```python
def expand_segments(segments, expansion, vert=True):
    
    paths = segments.get_paths()
    new_paths = []
    for i, path in enumerate(paths):
        if vert:
            offset = np.array([i+1, 0])
            new_path = (path.vertices-offset)*np.array([expansion[i],1])+offset
            new_paths.append(new_path)
        else:
            pass
    
    segments.set_paths(new_paths)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
results = ax.violinplot(vote_share_sen10.T)

for pc in results['bodies']:
    pc.set_facecolor('red')
    pc.set_edgecolor('black')
    pc.set_alpha(0.9)


paths = results['cmaxes'].get_paths()
new_paths = []
for i, path in enumerate(paths):
    offset = np.array([i+1, 0])
    new_path = (path.vertices-offset)*np.array([3,1])+offset
    new_paths.append(new_path)
results['cmaxes'].set_paths(new_paths)

plt.show()
```
